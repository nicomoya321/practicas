```python
#Correlaciones
plt.figure(dpi = 120,figsize= (6,4))
mask = np.triu(np.ones_like(public.corr(),dtype = bool))
sns.heatmap(public.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 90)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Importar librerias
import pandas as pd
import numpy as np

import seaborn as sns
import plotly.express as px
```

```python
arr= np.random.randn(50,4)
public= pd.DataFrame(arr, columns=['A','B','C','D'])
public.head()
public.plot()
```

```python
import plotly.graph_objects as go 
```

```python
colors=['blue','red','black']
fig= go.Figure(data=[go.Pie(labels=['TIPO_TRANSPORTE','CANTIDAD','DIA'],
                            values=[110,90,80,80,70,60])])
fig.update_traces(hoverinfo='label+percent', textfont_size=20,
                  textinfo= 'label+percent',pull=[0.1, 0, 0.2, 0, 0, 0],
                  marker=dict(colors=colors, line=dict(color='#FFFFFF',
                                                       width=2)))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=public, y=public,
                         mode='lines', name='colectivo'))
fig.add_trace(go.Scatter(x=public, y=public,
                         mode='lines+markers', name='subte'))
fig.add_trace(go.Scatter(x=public, y=public,
                         mode='lines+markers', name='trenes',
                         line=dict(color='firebrick',width=2,
                                   dash='dashdot')))
fig.update_layout(title='Precio de boleto 20/22',
                  xaxis_title='Precio',yaxis_title='Fecha')
```
